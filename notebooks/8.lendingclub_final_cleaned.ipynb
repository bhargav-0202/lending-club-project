```python
customer_bad_data_df =bad_data_customer_df.select("member_id") \
.union(bad_data_loans_defaulters_delinq_df.select("member_id")) \
.union(bad_data_loans_defaulters_detail_records_enquire_df.select("member_id"))

# this customer_bad_data_df will have all the records from three df's, repeated member_id's
```

```python
bad_customer_data_final_df = customer_bad_data_df.distinct()
```

## Segregating the Identified Bad data from the Normal Data

```python
bad_customer_data_final_df.createOrReplaceTempView("bad_data_customers")
```

```python
customers_df = spark.sql("""select * from itv022692_lending_club.customers where member_id NOT in (select member_id from bad_data_customers)""")
```

```python
customers_df.write \
.format("parquet") \
.mode("overwrite") \
.option("path","/user/itv022692/lendingclubproject/cleaned_new/customers_parquet") \
.save()
```

```python
loans_defaulters_delinq_df = spark.sql("""select * from itv022692_lending_club.loans_defaulters_delinq where member_id NOT in (select member_id from bad_data_customers)""")
```

```python
loans_defaulters_delinq_df.write \
.format("parquet") \
.mode("overwrite") \
.option("path","/user/itv022692/lendingclubproject/cleaned_new/loan_defaulters_delinq_parquet") \
.save()
```

```python
loans_defaulters_detail_records_enquire_df = spark.sql("""select * from itv022692_lending_club.loans_defaulters_detail_records_enquire where member_id NOT in (select member_id from bad_data_customers)""")
```

```python
loans_defaulters_detail_records_enquire_df.write \
.format("parquet") \
.mode("overwrite") \
.option("path","/user/itv022692/lendingclubproject/cleaned_new/loans_defaulters_detail_records_enquire_parquet") \
.save()
```
